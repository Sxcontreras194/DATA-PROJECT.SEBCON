```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#To see the first five in the data said, .tail for the last 5
df.head()
```

```python
# .sample gives you random samples within the data

df.sample(10)
```

```python
# setting random_state to an integer will allow you to get a random sample that you can then call back when you use that integer, in this case "42"
df.sample(10, random_state=42)
```

```python
# .copy makes a copy of the objects indices and data

dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()

#above we created a new dataset called df_original
```

```python
df_altered = df_original.copy()
```

```python
df_altered.loc[:5, 'salary_year_avg']
```

```python
median_salary = df_altered['salary_year_avg'].median()

median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

df_altered.loc[:5, 'salary_year_avg']

#The NaN values are now filled in with the median
```

```python
df_original.loc[:5, 'salary_year_avg']

#the original dataframe also got updated, this is not a good thing! This is because df_altered was set equal to df_original, now we updated it using .copy
```

```python
# by using .copy we don't have to do what we did above, other datasets won't be affected
# the .copy method should be used anytime we need to create a copy of the dataframe to protect us from causing harm anywhere else
```
